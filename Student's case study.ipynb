```python
# Building a regression model
import statsmodels.api as sm

X = students[['Attendance', 'HW', 'Test1', 'Project1', 'Test2', 'Project2']]
y = students['Final']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```
